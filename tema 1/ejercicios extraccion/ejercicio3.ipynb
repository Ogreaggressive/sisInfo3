# Extraer en un archivo CSV los siguientes datos de las noticias del autor 'Swati Dubey': fecha, hora, título y url

```python
#https://inshorts.deta.dev/news?category=science
import requests
respuesta = requests.get("http://localhost:8000/inshort-news.json")
data = respuesta.json()
data
```

```python
from jsonpath_ng import jsonpath
from jsonpath_ng.ext import parse
```

```python
autor = [match.value for match in parse('$..author').find(data)]
```

```python
autor
```

```python
datos = [match.value for match in parse('$.data[*]').find(data)]
datos
```

```python
jsonpath_expr = parse('$.data[*].author')
[match.value for match in jsonpath_expr.find(data)]
```

```python
jsonpath_expr = parse('$.data[?(@.author =~ "Swati Dubey")]')
```

```python
[match.value for match in jsonpath_expr.find(data)]
```

```python
matches = jsonpath_expr.find(data)
```

```python
matches[0].value
```

```python
matches[0].value["title"]
```

```python
fechas = [match.value["date"] for match in jsonpath_expr.find(data)]
horas = [match.value["time"] for match in jsonpath_expr.find(data)]
titulo = [match.value["title"] for match in jsonpath_expr.find(data)]
url = [match.value["url"] for match in jsonpath_expr.find(data)]
```

```python
tuples = list(zip(fechas, horas, titulo, url))
records = [{"date":t[0], "time": t[1], "title": t[2], "url": t[3]} for t in tuples]
```

```python
import csv

with open("Swati Dubey datos.csv", "w", encoding="utf-8", newline="") as f:
    writer = csv.DictWriter(f, fieldnames=['date', 'time', 'title', 'url'])
    writer.writeheader()
    writer.writerows(records)
```
