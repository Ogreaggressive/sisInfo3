**Extraer la masa, densidad, gravedad y duración del día de la siguiente fuente**

````python
import requests

response = requests.get("http://localhost:8000/Planetary_Fact_Sheet.html")
response
````

````python
from bs4 import BeautifulSoup
pagina = BeautifulSoup(response.text, "html.parser" )

MainTable = pagina.body.find_all(name="table",attrs={"role":"presentation"})
MainTable=MainTable[0]
TableRowsArray=MainTable.find_all(name="tr")
TableRowsArray
````

````python
#encontrar los id con align left
element_tds = pagina.body.find_all(name="td", attrs={"align":"left"})
element_tds[0]
````

````python
#Observando el archivo vemos que las filas que nos interesan son la 0 1 3 4 7

def getPlanetNames(PlanetsRow):
    Planets=[]
    for td in PlanetsRow.find_all(name="td"):
        anchorValue=td.find(name="a")
        if anchorValue:
            Planets.append(anchorValue.text)
    return Planets
def getDataFromNotHeaderRowFloats(Row):
    AnsArray=[]
    for td in Row.find_all(name="td"):
        anchorValue=td.find(name="a")
        if not anchorValue:
            AnsArray.append(float(td.text))
    return AnsArray

def getDataFromTable(TableRowsArray,DataIndexes):
    Planets=getPlanetNames(TableRowsArray[0])
    ValuesArray=[]  #al final tendra masa,densidad, gravedad,duracion del dia
    for index in DataIndexes:
        ValuesArray.append(getDataFromNotHeaderRowFloats(TableRowsArray[index]))
    return Planets,ValuesArray
DataIndexes=set([1,3,4,7])
Planets,Values=getDataFromTable(TableRowsArray,DataIndexes) #Values [[...masa],[...densidad],[...gravedad],[...duracion]]
print(Planets)
print(Values)
            
````


**Mostrar los datos de los tres planetas con mayor duración del día que La Tierra ordenados de manera descendente (por duración del día)**

````python
TuplePlanetsARRAY=[(Planets[index],Values[0][index],Values[1][index],Values[2][index],Values[3][index]) for index in range(len(Planets))]
TuplePlanetsARRAY #(NombrePlaneta,masa,densidad,gravedad,Duracion)

def sortSubject(PlanetTuple):
    return PlanetTuple[4]
TuplePlanetsARRAY.sort(key=sortSubject,reverse=True)
header = ["masa", "densidad", "gravedad", "duracion_dia"]
print(header)
print(TuplePlanetsARRAY[:3])
````

**Guardar en un archivo CSV los siguientes datos: el nombre, número de teléfono y rating de todos los negocios con ```"price"="$"```**

````python
import requests

response = requests.get("http://localhost:8000/businesses.json")
response
````

````python
from jsonpath_ng import jsonpath
from jsonpath_ng.ext import parse
````

````python

data = response.json()
data
````

````python
import csv
def saveToCSV(Header,Rows,CSVName):
    with open(CSVName, "w", encoding="utf-8", newline="") as f:
        writer = csv.DictWriter(f, fieldnames=Header)
        writer.writeheader()
        writer.writerows(Rows)
````

````python
jsonpath_expr = parse('$.businesses[?(@.price == "$")]')
[match.value for match in jsonpath_expr.find(data)]
````

````python
nombre = [match.value["name"] for match in jsonpath_expr.find(data)]
telefono = [match.value["phone"] for match in jsonpath_expr.find(data)]
rating = [match.value["rating"] for match in jsonpath_expr.find(data)]
````

````python
tuples = list(zip(nombre,telefono,rating))
records = [{"name":t[0], "phone": t[1], "rating": t[2]} for t in tuples]
````

````python
saveToCSV(records[0].keys(),records,"Cut Rate stuff.csv")
````
