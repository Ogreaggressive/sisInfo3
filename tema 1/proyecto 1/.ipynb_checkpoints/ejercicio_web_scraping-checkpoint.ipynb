```python
#Importar la clase BeautifulSoup del módulo bs4
from bs4 import BeautifulSoup
pagina = BeautifulSoup(respuesta.text, "html.parser" )
# pagina?
# print(pagina.prettify())
#Demo. Chrome Developer Tools: Explorar la estructura de la página para identificar los elementos que contienen los datos
#https://www.tutorialspoint.com/beautiful_soup/beautiful_soup_encoding.htm
```

```python
print(pagina.prettify())
```

```python
name_divs = pagina.body.find_all(name="div", attrs={"class":"entry"})
div = name_divs[0]
div2 = name_divs[1]
print(div['class'])
print(div.prettify())
```

```python
text = div.text
text
text = text.strip()
text
titulo, texto = text.split('(')
titulo
titulo.strip()
```

```python
def extraer_nombre(div_class_entry):
    text = div_class_entry.text
    text = text.strip()
    titulo, texto = text.split('(')
    titulo.strip()
    return titulo
```

```python
extraer_nombre(div)
```

```python
name_divs = pagina.body.find_all(name="div", attrs={"class":"entry"})
filas = []
for name_div in name_divs:
    fila = extraer_nombre(name_div)
    filas.append(fila)
print(f"Se extraido {len(filas)} filas")
print(filas)
```
